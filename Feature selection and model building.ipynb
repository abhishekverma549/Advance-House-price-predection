```python
import pandas as pd
import numpy as np
import pprint

## for feature slection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
#train_data = pd.read_csv(r'C:\Users\abhiv\Desktop\DS\Data Science Projects\Advance House price predection\Cleaned data\Train_data.csv')
#test_data = pd.read_csv(r'C:\Users\abhiv\Desktop\DS\Data Science Projects\Advance House price predection\Cleaned data\Test_data.csv')
```

```python
train_data = pd.read_csv('./Cleaned data/{}'.format('Train_data.csv'))
test_data = pd.read_csv('./Cleaned data/{}'.format('Test_data.csv'))
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
X_train = train_data.drop(['Id', 'SalePrice'], axis=1)
y_train = train_data[['SalePrice']]
```

```python
X_test = test_data.drop('Id', axis=1)
```

```python
print('X train data shape {}'.format(X_train.shape))
print('y train data shape {}'.format(y_train.shape))
print('test data shape {}'.format(X_test.shape))
```

# Feature selection for training data

```python
feature_sel_model_train = SelectFromModel(Lasso(alpha=0.004, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model_train.fit(X_train, y_train)
```

```python
feature_sel_model_train.get_support()
```

```python
selected_feature = X_train.columns[feature_sel_model_train.get_support()]


print('Total no of feature {}'.format(X_train.shape[1]))
print('total Selected features {}'.format(len(selected_feature)))
print('feature with coef shrank to zero {}'.format(np.sum(feature_sel_model_train.estimator_.coef_ == 0)))
```

```python
X_train = X_train[selected_feature]
X_train.head()
```

# Feature selection for test data

```python
X_test = X_test[selected_feature]
X_test.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
forest_model = RandomForestRegressor()
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 'log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
#max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
pprint.pprint(random_grid)
```

```python
random_search = RandomizedSearchCV(estimator = forest_model, param_distributions = random_grid, n_iter = 100, cv = 5, verbose =2, random_state = 42, n_jobs = -1)
```

```python
#Fit the random search model
random_search.fit(X_train, y_train)
```

```python
random_search.best_params_
```

```python
random_search.best_estimator_
```

```python
forest_model = RandomForestRegressor(bootstrap=False, max_depth=100, max_features='sqrt',
                      n_estimators=800, n_jobs=-1)
```

```python
forest_score = cross_val_score(forest_model,X_train, y_train, cv=5)
print('Score of 5 Kfold CV :',forest_score)
print('\nAverage CV score: ',np.round(forest_score.mean()*100, 4))
```

```python
forest_model.fit(X_train, y_train)
score_RF = forest_model.score(X_train, y_train)
print('R Squared :', np.round(score_RF*100, 2))

#y_pred_RF = forest_model.predict(X_test)
```

```python
y_pred_RF = forest_model.predict(X_test)
```

```python
pred = pd.DataFrame(np.exp(y_pred_RF))
sub_df = pd.read_csv('./Raw data/{}'.format('sample_submission.csv'))
dataset = pd.concat([sub_df['Id'], pred], axis = 1)
```

```python
dataset.columns = [['Id', 'SalePrice']]
```

```python
dataset.to_csv('./Cleaned data/{}'.format('submission_RF.csv'), index= False)
```

```python
xgboost = XGBRegressor()
```

```python
n_estimators = [int(x) for x in np.linspace(200, 2000, 10)]
max_depth = [int(x) for x in np.linspace(10, 100, 10)]
learning_rate=[0.05,0.1,0.15,0.20]
booster = ['gbtree', 'gblinear', 'dart']
min_child_weight=[int(x) for x in np.linspace(1, 10, 10)]
```

```python
parameters = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'learning_rate': learning_rate,
    'booster': booster,
    'min_child_weight': min_child_weight,
}
```

```python
random_cv = RandomizedSearchCV(estimator = xgboost,
                               param_distributions = parameters,
                               n_iter = 50, cv = 5,
                               n_jobs = -1, verbose = 5,
                               random_state = 42,
                               return_train_score = True)
```

```python
random_cv.fit(X_train, y_train)
```

```python
random_cv.best_params_
```
